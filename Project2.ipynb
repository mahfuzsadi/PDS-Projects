```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1. Create two CSV files with test results
students = 50
np.random.seed(42)

data1 = {
    "Registration Number": range(1, students + 1),
    "Name": [f"Student_{i}" for i in range(1, students + 1)],
    "TT-1 Marks": np.random.randint(0, 20, students)  # Random marks out of 20
}

data2 = {
    "Registration Number": range(1, students + 1),
    "Name": [f"Student_{i}" for i in range(1, students + 1)],
    "TT-2 Marks": np.random.randint(0, 20, students)  # Random marks out of 20
}
```

```python
df1 = pd.DataFrame(data1)
df1
```

```python
df2 = pd.DataFrame(data2)
df2
```

```python
df1.to_csv("term-test-1-result.csv", index=False)
df2.to_csv("term-test-2-result.csv", index=False)
```

```python
df1 = pd.read_csv("term-test-1-result.csv")
df2 = pd.read_csv("term-test-2-result.csv")
df1.head()
```

```python
df_merged = pd.merge(df1, df2, on=["Registration Number", "Name"])
df_merged.head()
```

```python
df_merged["Best TT Marks"] = df_merged[["TT-1 Marks", "TT-2 Marks"]].max(axis=1)
df_merged["Avg TT Marks"] = df_merged[["TT-1 Marks", "TT-2 Marks"]].mean(axis=1)
df_merged
```

```python
attendance_marks = np.random.randint(5, 10, students)  # Random attendance marks (out of 10)
final_term_marks = np.random.randint(40, 100, students)  # Random final term marks (out of 100)

df_merged["Attendance Marks"] = attendance_marks
df_merged["Term Final Marks"] = final_term_marks
df_merged.head()
```

```python
df_merged["Final Marks"] = (df_merged["Term Final Marks"] * 0.7) + df_merged["Avg TT Marks"] + df_merged["Attendance Marks"]

# Save final results to CSV
df_merged.to_csv("final-result.csv", index=False)
df_merged.head()
```

```python
X = df_merged[["Final Marks"]].values
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_merged["Cluster"] = kmeans.fit_predict(X)
df_merged.head()
```

```python
plt.scatter(df_merged["Registration Number"], df_merged["Final Marks"], c=df_merged["Cluster"], cmap="viridis", edgecolors="k")
plt.xlabel("Registration Number")
plt.ylabel("Final Marks")
plt.title("K-Means Clustering of Students Based on Final Marks")
plt.colorbar(label="Cluster")
plt.show()
```
